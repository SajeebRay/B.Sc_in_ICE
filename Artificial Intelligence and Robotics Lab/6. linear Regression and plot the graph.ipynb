# Implement simple linear Regression and plot the graph

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('1.car driving risk analysis.csv')
df.head()
```

```python
x=df[['speed']]  
y=df[['risk']]    
x.head()
```

```python
y.head()
```

# Visualization


```python
plt.scatter(x,y)
plt.xlabel('Speed of Car')
plt.ylabel('Risk on driving')
plt.title('Car driving speed risk')
```

# Risk should be analysed based on speed
#60% For training and 40% for test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.40,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
plt.scatter(x,y,marker='*',color='red')
plt.xlabel('Speed of Car')
plt.ylabel('Risk on driving')
plt.title('Car driving speed risk')
plt.plot(x,model.predict(x))
```

```python
coff = model.coef_
print("Coefficient: ", coff)
intercept = model.intercept_
print("intercept: ", intercept)
```

```python
y_pred = model.predict(xtest)  #compare with ytest
y_pred
```

User input speed

```python
risk = model.predict([[150]])
print("When speed 150, then risk is: ", risk[0,0])
```

```python
print("Model Score is: ")
model.score(xtest,ytest)
```
